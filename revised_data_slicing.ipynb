```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import os

# Constants
G_TO_MPS2 = 9.80665
MILLI_G_TO_MPS2 = G_TO_MPS2 / 1000
FORCE_THRESHOLD = 300
WINDOW_SIZE = 0.4

# Load and prepare the data
def load_data_from_directory(directory):
    datasets = []
    file_paths = []
    for filename in os.listdir(directory):
        if filename.endswith('.xlsx'):
            file_path = os.path.join(directory, filename)
            data = pd.read_excel(file_path)
            datasets.append(data)
            file_paths.append(file_path)
    return datasets, file_paths

# Function to find absolute top peaks
def find_absolute_top_peaks(data, min_time_separation=0.5, force_threshold=FORCE_THRESHOLD, num_peaks=5):
    peaks, _ = find_peaks(data['Total_GRF'], height=force_threshold)
    valid_data = data.iloc[peaks]
    sorted_data = valid_data.sort_values(by='Total_GRF', ascending=False)

    top_peaks = []
    for index, row in sorted_data.iterrows():
        if len(top_peaks) >= num_peaks:
            break

        time_diffs = [abs(row['Time'] - data['Time'].iloc[prev_index]) for prev_index in top_peaks]
        if all(diff >= min_time_separation for diff in time_diffs):
            top_peaks.append(index)

    return top_peaks

# Function to visualize peaks and windows
def visualize_peaks_and_windows(data, top_peak_indices):
    plt.figure(figsize=(15, 6))
    plt.plot(data['Time'], data['Total_GRF'], label='Total_GRF')

    for idx, peak_idx in enumerate(top_peak_indices, 1):
        peak_time = data['Time'].iloc[peak_idx]
        start_window = peak_time - WINDOW_SIZE / 2
        end_window = peak_time + WINDOW_SIZE / 2

        plt.plot(peak_time, data['Total_GRF'].iloc[peak_idx], 'X', markersize=12, label=f'Peak {idx}')
        plt.axvspan(start_window, end_window, color='red', alpha=0.2)

    plt.xlabel('Time (s)')
    plt.ylabel('Resultant Force (unit)')
    plt.title('Absolute Top Force Peaks with Event Windows')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to generate summary table
def generate_summary_table(data, top_peak_indices):
    summary = []
    for i, peak_idx in enumerate(top_peak_indices):
        event_data = data.iloc[peak_idx]
        time_of_max_force = event_data['Time']
        peak_force_value = event_data['Total_GRF']
        resultant_acceleration = event_data['Total_GRF']

        summary.append({
            'Event': i + 1,
            'Time of Maximum Force (s)': time_of_max_force,
            'Peak Force Value (unit)': peak_force_value,
            'Resultant Acceleration (m/s^2)': resultant_acceleration
        })

    summary_table = pd.DataFrame(summary)
    return summary_table

# Function to extract and save events
def extract_and_save_events(data, top_peaks, original_file_path, window_size=WINDOW_SIZE):
    base_name = original_file_path.rsplit('.', 1)[0]
    for i, peak_index in enumerate(top_peaks):
        peak_time = data['Time'].iloc[peak_index]
        start_time = peak_time - window_size / 2
        end_time = peak_time + window_size / 2
        event_data = data[(data['Time'] >= start_time) & (data['Time'] <= end_time)]

        event_file_name = f'{base_name}_event_{i + 1}.xlsx'        
        event_data.to_excel(event_file_name, index=False)
        print(f'Event {i + 1} data saved to {event_file_name}.')

# Main processing
directory = os.getcwd()
datasets, file_paths = load_data_from_directory(directory)

for data, file_path in zip(datasets, file_paths):
    data[['2x', '2y', '2z', '3x', '3y', '3z', '4x', '4y', '4z']] *= MILLI_G_TO_MPS2
    data['resultant_acceleration_1'] = np.sqrt(data['2x']**2 + data['2y']**2 + data['2z']**2)
    data['resultant_acceleration_2'] = np.sqrt(data['3x']**2 + data['3y']**2 + data['3z']**2)
    data['resultant_acceleration_3'] = np.sqrt(data['4x']**2 + data['4y']**2 + data['4z']**2)

    top_peak_indices = find_absolute_top_peaks(data)
    visualize_peaks_and_windows(data, top_peak_indices)
    summary_table = generate_summary_table(data, top_peak_indices)
    print(summary_table)
    extract_and_save_events(data, top_peak_indices, file_path)
```
